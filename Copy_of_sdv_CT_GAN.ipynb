```python
from sdv.tabular import CTGAN
import pandas as pd

data = pd.read_csv("Threat-Actor-Langchain.csv")

model = CTGAN()

model.fit(data)
```

```python
model.save('Threat_Actor_CTGAN.pkl')

loaded = CTGAN.load('Threat_Actor_CTGAN.pkl')

new_data = loaded.sample(num_rows=1000000)
```

```python
import json
import pandas as pd  # Assuming your model output is easily loaded into a Pandas DataFrame

# ----------------
# 1. Load Model Output
# ----------------
model_output = pd.read_csv('new_data.csv')  # Replace with the appropriate method

# ----------------
# 2. Data Mapping
# ----------------

def map_row_to_stix(row):
    """Maps a row of your model output to a STIX object dictionary"""

    stix_object = {
        "type": "threat-actor",
        "id": f"threat-actor--{uuid.uuid4()}",  # Generate unique STIX ID
        "spec_version": "2.1",
        "name": row.get('name', ''),  # Get 'name' if it exists
        "description": row.get('description', ''),
        "threat_actor_types": row.get('threat_actor_types', []),  # Assuming this is a list in your data
        "aliases": row.get('aliases', []),
        "roles": row.get('roles', []),
        "goals": row.get('goals', []),
        "sophistication": row.get('sophistication', ''),
        "resource_level": row.get('resource_level', ''),
        "primary_motivation": row.get('primary_motivation', '')
    }
    return stix_object

# ----------------
# 3. Build STIX Bundle
# ----------------

stix_objects = model_output.apply(map_row_to_stix, axis=1).tolist()

stix_bundle = {
    "type": "bundle",
    "spec_version": "2.1",
    "objects": stix_objects
    # Optionally add relationships if your model provides that data
}

# ----------------
# 4. Save as JSON
# ----------------

with open('stix_data.json', 'w')  as f:
    json.dump(stix_bundle, f, indent=2)
```
